<font color='darkred'>Ο αλγόριθμος Bellman-Ford είναι ένας αλγόριθμος εύρεσης της συντομότερης διαδρομής (shortest path) από έναν κόμβο ενός γράφου προς όλους τους υπόλοιπους κόμβους του γράφου. Ο γράφος αυτός είναι σταθμισμένος (weighted) που σημαίνει ότι κάθε του ακμή έχει ένα βάρος (weight), το οποίο καλείται και κόστος. Ο αλγόριθμος μπορεί να βρει την συντομότερη διαδρομή ακόμα και αν υπάρχουν αρνητικά βάρη σε ορισμένες ακμές, σε σχέση με άλλους αλγορίθμους εύρεσης της συντομότερης διαδρομής, όπως του Dijkstra που απαιτεί όλες οι ακμές να έχουν μη αρνητικά βάρη. 


*  Ο αλγόριθμος ξεκινάει θέτοντας την απόσταση του κόμβου πηγή σε 0 και την απόσταση όλων των υπόλοιπων κόμβων σε ∞
*   Ο αλγόριθμος επαναλαμβάνεται σε όλες τις ακμές του γράφου $V-1$ φορές, όπου $V$ είναι ο αριθμός των κόμβων του γράφου
*   Σε κάθε επανάληψη «χαλαρώνει» κάθε ακμή του γράφου, ενημερώνοντας την απόσταση του κάθε κόμβου αν βρεθεί ένα συντομότερο μονοπάτι
*   Η διαδικασία «χαλάρωσης» μιας ακμής περιλαμβάνει τη σύγκριση της απόστασης από τον κόμβο πηγή συν το βάρος της ακμής με την τρέχουσα εκτίμηση της απόστασης από τον κόμβο πηγή.
*  Εάν η νέα απόσταση είναι μικρότερη, η εκτίμηση ενημερώνεται. </font>

<font color='darkred'>Κρίνεται σκόπιμο να αναφερθεί πως ο αλγόριθμος δεν τερματίζει αν ο γράφος περιέχει κάποιον αρνητικό κύκλο. Σε μία τέτοια περίπτωση αν πραγματοποιηθεί μία ακόμη επανάληψη μετά τις $V-1$ θα εντοπιστεί ένα συντομότερο μονοπάτι, το οποίο δηλώνει πως υπάρχει αρνητικός κύκλος. Με την ύπαρξη ενός αρνητικού κύκλου το συντομότερο μονοπάτι μειώνεται επ’ αόριστον και το πρόβλημα δεν έχει λύση (unbounded).

Για την επίλυση της άσκησης θα χρησιμοποιήσετε την παρακάτω υλοποίηση του αλγορίθμου *Bellman Ford*:

```
# Python3 program for Bellman-Ford's single source
# shortest path algorithm.
# From: https://github.com/arnab132/Bellman-Ford-Algorithm-Python/blob/main/bellman_ford.py

# Class to represent a graph
class Graph:

	def __init__(self, vertices):
		self.V = vertices # No. of vertices
		self.graph = []

	# function to add an edge to graph
	def addEdge(self, u, v, w):
		self.graph.append([u, v, w])
		self.graph.append([v, u, w])
		
	# utility function used to print the solution
	def printArr(self, dist):
		print("Vertex Distance from Source")
		for i in range(self.V):
			print("{0}\t\t{1}".format(i, dist[i]))
	
	# The main function that finds shortest distances from src to
	# all other vertices using Bellman-Ford algorithm. The function
	# also detects negative weight cycle
	def BellmanFord(self, src):

		# Step 1: Initialize distances from src to all other vertices
		# as INFINITE
		dist = [float("Inf")] * self.V
		dist[src] = 0


		# Step 2: Relax all edges |V| - 1 times. A simple shortest
		# path from src to any other vertex can have at-most |V| - 1
		# edges
		for _ in range(self.V - 1):
			# Update dist value and parent index of the adjacent vertices of
			# the picked vertex. Consider only those vertices which are still in
			# queue
			for u, v, w in self.graph:
				if dist[u] != float("Inf") and dist[u] + w < dist[v]:
						dist[v] = dist[u] + w

		# Step 3: check for negative-weight cycles. The above step
		# guarantees shortest distances if graph doesn't contain
		# negative weight cycle. If we get a shorter path, then there
		# is a cycle.

		for u, v, w in self.graph:
				if dist[u] != float("Inf") and dist[u] + w < dist[v]:
						print("Graph contains negative weight cycle")
						return
						
		# print all distance
		self.printArr(dist)
```

Ένα παράδειγμα ορισμού γράφου για να ξεκινήσετε είναι το παρακάτω (προσοχή στην αρίθμηση των κόμβων):

```
g = Graph(5) # number of nodes (Must include node 0)
g.addEdge(0, 1, -1)
g.addEdge(0, 2, 4)
g.addEdge(1, 2, 3)
g.addEdge(1, 3, 2)
g.addEdge(1, 4, 2)
g.addEdge(3, 2, 5)
g.addEdge(3, 1, 1)
g.addEdge(4, 3, -3)

# Print the solution
g.BellmanFord(0)
```
