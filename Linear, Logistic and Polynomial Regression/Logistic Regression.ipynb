<font color='darkred'>Η Λογιστική Παλινδρόμηση είναι τεχνική ταξινόμησης, η οποία ανήκει στην ομάδα των γραμμικών ταξινομητών και χρησιμοποιείται κατά κόρον σε προβλήματα δυαδικής ταξινόμησης, όπως στην συγκεκριμένη άσκηση. Εφαρμόζεται για την πρόβλεψη μίας κατηγορικής εξαρτημένης μεταβλητής από ένα σύνολο ανεξάρτητων μεταβλητών.</font>

<font color='darkred'>Στην Λογιστική Παλινδρόμηση ως συνάρτηση ενεργοποίησης χρησιμοποιείται η σιγμοειδής που δίνεται από τον τύπο f(x) = 1/(1+ e^(-x)). H σιγμοειδής συνάρτηση μετατρέπει οποιαδήποτε πραγματική τιμή σε εύρος από 0 έως 1. Αν η τιμή της σιγμοειδής είναι μεγαλύτερη του 0.5 τότε το ο ταξινομητής προβλέπει ότι η το δείγμα ανήκει στην κλάση 1, ενώ διαφορετικά στην κλάση 0. </font>

<font color='darkred'>Στην συγκεκριμένη άσκηση θα κατασκευαστεί ένας ταξινομητής Spam μηνυμάτων χρησιμοποιώντας την Λογιστική Παλινδρόμηση. Συγκεκριμένα θα χρησιμοποιήσουμε ένα σύνολο από SMS προσπαθώντας να κατηγοριοποιήσουμε αν είναι spam (κλάση 1) ή όχι (κλάση 0). Το dataset περιέχει 4.827 SMS με την ονομασία ham, τα οποία είναι ασφαλή (86.6%) και 747 SMS, τα οποία είναι spam (13.4%). </font>

<font color='darkred'>Σαν πρώτο βήμα διαβάζουμε τα δεδομένα από το αρχείο SMSSpamCollection.csv, στο οποίο δεν υπάρχει επικεφαλίδα, ενώ επιλέγουμε να ονομάσουμε τις στήλες Class Label και SMS. Στην στήλη Class Label αντικαθιστούμε τα ham(ασφαλή) SMS με 0 και τα spam με 1. Με αυτό τον τρόπο έχουμε ένα πρόβλημα δυαδικής ταξινόμησης.


```
# διάβασμα δεδομένων
import pandas as pd
filename = 'SMSSpamCollection.csv'
df = pd.read_csv(filename, sep = '\t', header = None, names = ['Class Label', 'SMS'])
df = df.replace(['ham','spam'],[0,1])
df.head()
```

```
HAM = df[df['Class Label']==0].count()[0]
SPAM = df[df['Class Label']==1].count()[0]
print(f"Το {100*HAM/df.shape[0]:.2f}% των SMS αντιστοιχεί σε ham μηνύματα.")
print(f"Το {100*SPAM/df.shape[0]:.2f}% των SMS αντιστοιχεί σε spam μηνύματα.")
```

<font color='darkred'> Χωρίζουμε τα δεδομένα με τυχαίο τρόπο σε δεδομένα εκπαίδευσης (training set) και δεδομένα για τον υπολογισμό της ακρίβειας του μοντέλου (test set). Εφόσον δεν δίνεται κάποια οδηγία χρησιμοποιώ την αναλογία 70% train και 30% test. Επίσης, το random_state = 30 χρησιμοποιείται ώστε να έχουμε κάθε φορά τα ίδια αποτελέσματα. 

```
# Διαχωρισμός dataset 70-30
from sklearn.model_selection import train_test_split
X_train_raw, X_test_raw, y_train, y_test = train_test_split(df['SMS'], df['Class Label'], random_state=30, test_size=0.3)
```

<font color='darkred'>Στην συνέχεια κάνουμε προεπεξεργασία των δεδομένων χρησιμοποιώντας τη μέθοδο TfidVectorizer της βιβλιοθήκης Scikit-Learn. Η μέθοδος αυτή μετατρέπει ακατέργαστα δεδομένα κειμένου σε πίνακα TF-IDF (Term Frequency -Inverse Document Frequency), ο οποίος φανερώνει την σημασία κάθε λέξης στο κείμενο. Αξίζει να σημειωθεί πως η μέθοδος TfidVectorizer  αφαιρεί και τα σημεία στίξης καθώς επεξεργάζεται το dataset.

```
# προεπεξεργασία δεδομένων
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train_raw)
X_test = vectorizer.transform(X_test_raw)
```

<font color='darkred'>Τώρα είμαστε σε θέση να εκπαιδεύσουμε ένα μοντέλο Λογιστικής Παλινδρόμησης στα δεδομένας μας.

```
# Εκπαίδευση μοντέλου Λογιστικής Παλινδρόμησης
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, y_train)
```

<font color='darkred'>Για να κρίνουμε την αποτελεσματικότητα του μοντέλου μπορούμε να υπολογίσουμε την ακρίβεια του πάνω στο test test, η οποία όπως βλέπουμε παρακάτω είναι πολύ υψηλή και ίση με περίπου 97.25%

```
from sklearn.metrics import accuracy_score 
predictions = logisticRegr.predict(X_test)
print("Η Aκρίβεια του μοντέλου είναι ίση με:", accuracy_score(y_test, predictions))
```

<font color='darkred'>Ωστόσο, στη συγκεκριμένη άσκηση που ο στόχος είναι να εντοπίσουμε τα spam μηνύματα ένα καλύτερο μέτρο είναι το F1- Score, το οποίο είναι ο αρμονικός μέσος όρος του Precision και Recall. Αυτό το μέτρο είναι καλύτερο, καθώς τα spam μηνυμάτα είναι πολύ λιγότερα σε αναλογία με τα ham. Το F1-Score προκύπτει ίσο με περίπου 87.89%, το οποίο επίσης είναι πολύ υψηλό. Τέλος, μπορούμε να τυπώσουμε και το classification report που μας δίνει περισσότερες πληροφορίες υπολογίζοντας Precision, Recall, F1-Score, Accuracy, F1-Macro και F1-Weighted.

```
from sklearn.metrics import f1_score #f1score
from sklearn.metrics import classification_report #classification report
predictions = logisticRegr.predict(X_test)
print("Το F1-Score του μοντέλου είναι ίσο με:",f1_score(y_test, predictions))
print("Το Classification Report είναι: \n",classification_report(y_test, predictions))
```

Στην γραμμική παλινδρόμηση η έξοδος που μοντελοποιείται είναι μία αριθμητική τιμή ενώ μία βέλτιστη ευθεία προσαρμόζεται στα δεδομένα.
