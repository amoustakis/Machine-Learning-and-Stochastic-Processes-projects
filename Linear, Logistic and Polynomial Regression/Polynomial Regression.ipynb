```
!pip install numpy
!pip install matplotlib
!pip install pandas
!pip install sklearn

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

Στη συνέχεια, θα φορτώσετε τα δεδομένα εκπαίδευσης που παρέχονται από το αρχείο
 <b><i>data2.csv</b></i>.

```
data = pd.read_csv('data2.csv')
print(data)
```

Τώρα, θα δημιουργήσετε τις μεταβλητές εισόδου και εξόδου για τους αλγορίθμους μηχανικής μάθησης από τα δεδομένα εκπαίδευσης που παρέχονται.

```
X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values

print("Input")

print(X)

print("Labels")

print(y)
```

Στη συνέχεια, θα εκπαιδεύσετε το μοντέλο <b><i>linear regression</i></b>, χρησιμοποιώντας τα παραπάνω δεδομένα εκπαίδευσης.

```
lin = LinearRegression()
lin.fit(X, y)
```

**<h4> <b><i>degree</b></i> = 2 </h4>**

```
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

<font color='darkred'>Η Πολυωνυμική Παλινδρόμηση είναι μία μορφή γραμμικής παλινδρόμησης στην οποία η σχέση μεταξύ της ανεξάρτητης μεταβλητής x και της εξαρτημένης μεταβλητής y μοντελοποιείται ως πολυώνυμο n βαθμού, δηλαδή $y = a + b_1x +b_2x^2+ … +b_n x^n$, όπου $n$ ο βαθμός (degree) του πολυωνύμου.</font>

<font color='darkred'>Καθώς αυξάνεται η τιμή της παραμέτρου degree, δηλαδή ο βαθμός του πολυωνύμου, παρατηρούμε πως η γραμμή παλινδρόμησης προσαρμόζεται καλύτερα πάνω στα δεδομένα εκπαίδευσης. Ωστόσο, αυτό ενέχει τον κίνδυνο της υπερπροσαρμογής του μοντέλου στα δεδομένα, το οποίο σημαίνει πως το μοντέλο προσαρμόζεται πολύ καλά στα δεδομένα εκπαίδευσης αλλά έχει μειωμένη γενίκευση (χαμηλές επιδόσεις στα δεδομένα ελέγχου). Επίσης, όσο μεγαλύτερος είναι ο βαθμός του πολυωνύμου τόσο πιο ευαίσθητο είναι το μοντέλο σε σημεία επιρροής/outliers.</font>

<font color='darkred'>Για τον υπολογισμό του χρόνου εκπαίδευσης για τις διάφορες εξεταζόμενες τιμές του degree παραθέτω το παρακάτω barplot , όπου στον οριζόντιο άξονα είναι οι βαθμοί του πολυωνύμου και στον κάθετο άξονα ο αντίστοιχος χρόνος εκπαίδευσης. Το barplot δημιουργήθηκε αφού πρώτα υπολόγισα τον χρόνο εκπαίδευσης για κάθε μοντέλο.</font>



 


```
import time 
```

```
degrees = [2,3,4,5,8,10,12,15]
train = []
for degree in degrees:
  poly = PolynomialFeatures(degree=degree)
  X_poly = poly.fit_transform(X)
  start_time = time.time() 
  poly.fit(X_poly, y) 
  train_time = time.time() - start_time #χρόνος εκπαίδευσης
  train.append(train_time)
```

```
plt.bar(degrees,train)
plt.ylabel('Χρόνος εκπαίδευσης')
plt.xlabel('Βαθμός Πολυωνύμου')
plt.title('Πολυωνυμική Παλινδρόμηση')
```

<font color='darkred'>Παρατηρούμε πως οι χρόνοι εκπαίδευσης σε όλες τις περιπτώσεις είναι αμελητέοι και πολύ κοντά στα 0.00025 δευτερόλεπτα, ανεξάρτητα από τον βαθμό του πολυωνύμου. Γενικότερα  θα περιμέναμε να αυξάνεται ο χρόνος εκπαίδευσης όσο αυξάνεται ο βαθμός του πολυωνύμου, καθώς υπάρχουν περισσότεροι παράμετροι να εκπαιδευτούν. <h3><b><i>Ερώτημα 2</b></i></h3>

Σας παρέχεται, επίσης, το αρχείο <b><i>data2b.csv</i></b>, το οποίο είναι το αρχείο <b><i>data2.csv</i></b> με μία παραπάνω εγγραφή, την <b><i>(9, 38, 0.3)</i></b>. Να εκτελέσετε το παρακάτω πρόγραμμα για την τιμή της παραμέτρου <b><i>degree = 15</i></b>. Πόσο μεγάλη είναι η αλλαγή που παρατηρείτε στη μορφή της γραμμής παλινδρόμησης σε σχέση με την αντίστοιχη περίπτωση του ερωτήματος 1; Τι συμπέρασμα μπορείτε να βγάλετε για τον αλγόριθμο <b><i>polynomial regression</i></b>;

```
data = pd.read_csv('data2b.csv')
print(data)

X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values

lin = LinearRegression()
lin.fit(X, y)

poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

<font color='darkred'>Το σημείο που προστίθεται αντιστοιχεί σε outlier καθώς η τιμή πίεσης που διαθέτει (0.3) είναι αρκετά μεγάλη σε σχέση με τα υπόλοιπα δεδομένα του δείγματος. Στην συγκεκριμένη περίπτωση που ο βαθμός του πολυωνύμου είναι 15 η γραμμή της πολυωνυμικής παλινδρόμησης επηρεάζεται αρκετά καθώς προσπαθεί να πλησιάσει το νέο αυτό σημείο, όπως φαίνεται και στο διάγραμμα. Αυτό φανερώνει το πρόβλημα υπερπροσαρμογής που παρατηρείται όταν έχουμε πολυώνυμο μεγάλου βαθμού. Στην αντίστοιχη περίπτωση του ερωτήματος 1 δεν υπήρχε κάποιο σημείο που να αποτελεί outlier και να επηρεάζει σε μεγάλο βαθμό την γραμμή παλινδρόμησης. Συμπερασματικά, ένα μοντέλο πολυωνυμικής παλινδρόμησης με μεγάλο βαθμό πολυωνύμου είναι αρκετά ευαίσθητο σε outliers.</font>
