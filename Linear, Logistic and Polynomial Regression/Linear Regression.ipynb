## <font color='darkred'>Πρώτο dataset (data1a.csv)

<font color='darkred'>Θα ξεκινήσουμε φορτώνοντας τα δεδομένα του πρώτου προβλήματος (data1a.csv). Κρατάμε μόνο τις στήλες views και likes, οι οποίες αποτελούν την εξαρτημένη y και την ανεξάρτητη μεταβλητή x αντίστοιχα, καθώς θέλουμε με βάση τον αριθμό των likes να προβλέψουμε τα views που θα πάρει ένα YouTube video. 

```
df = pd.read_csv("https://raw.githubusercontent.com/nkostopoulos/StochasticsLabPublic/master/lab1/data1a.csv")
df = df[['views','likes']]
df.head()
```

<font color='darkred'>Στην συνέχεια υπολογίζεται η συσχέτιση (correlation) ανάμεσα στην εξαρτημένη και την ανεξάρτητη μεταβλητή. Η συσχέτιση μπορεί να εφαρμοστεί ώστε να αποφανθεί αν ένα μοντέλο γραμμικής παλινδρόμησης είναι κατάλληλο για να εφαρμοστεί στα δεδομένα. Η συσχέτιση παίρνει τιμές που κυμαίνονται από -1 έως +1, με τιμές πιο κοντά στις άκρες να δηλώνουν υψηλή συσχέτιση (αρνητική ή θετική αντίστοιχα) και τιμές πιο κοντά στο 0 να δηλώνουν καμία συσχέτιση. Προφανώς, για να εφαρμοστεί επιτυχημένα ένα μοντέλο γραμμικής παλινδρόμησης θέλουμε οι δύο μεταβλητές να εμφανίζουν ισχυρή είτε θετική είτε αρνητική συσχέτιση, δηλαδή να είναι γραμμικά συσχετισμένες. 

```
cor = df['views'].corr(df['likes'])
print('Η συσχέτιση των δύο μεταβλητών είναι:', cor)
```

<font color='darkred'>Με μία πρώτη ματιά ένα μοντέλο γραμμικής παλινδρόμησης μπορεί να εφαρμοστεί στα δεδομένα, καθώς έχουν υψηλό συντελεστή συσχέτισης κοντά στο 1 (περίπου 0,839). Ύστερα, θα δημιουργήσουμε το διάγραμμα διασποράς των δύο μεταβλητών.  

```
plt.scatter(df['likes'], df['views'])
plt.title('Διάγραμμα Διασποράς')
plt.xlabel('Likes')
plt.ylabel('Views')
plt.show()
```

<font color='darkred'>Παρόλο που υπάρχει υψηλή συσχέτιση παρατηρούμε στο διάγραμμα πως αν φέρναμε μία νοητή ευθεία αρκετά σημεία θα ξέφευγαν. Ωστόσο, το διάγραμμα δεν είναι τόσο αντιπροσωπευτικό καθώς περιέχει 7.996 σημεία, όσες και τα δείγματα του dataset.

<font color='darkred'>Στην συνέχεια γίνεται κανονικοποίηση των δεδομένων με την μέθοδο Min-Max. Η κανονικοποίηση Min-Max κλιμακώνει τα χαρακτηριστικά γραμμικά στο διάστημα [0,1] αφαιρώντας την ελάχιστη τιμή και διαιρώντας με τη διαφορά της ελάχιστης από την μέγιστη τιμή. 
$$
x' = \frac{x - x_{min}}{x_{max} - x_{min}}
$$

```
from sklearn.preprocessing import MinMaxScaler

scaled_data = MinMaxScaler().fit_transform(df)
```

<font color='darkred'>Στην συνέχεια υποδιαιρούμε τα δεδομένα που παρέχονται με τυχαίο τρόπο σε δύο υποσύνολα: το training set που περιλαμβάνει τα δεδομένα που θα χρησιμοποιηθούν για την εκπαίδευση του μοντέλου και το test set που περιλαμβάνει τα δεδομένα που θα χρησιμοποιηθούν για τον υπολογισμό της ακρίβειας του μοντέλου. Το training set είναι το 80% του αρχικού dataset και το test set είναι το 20%.

```
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest = train_test_split(scaled_data[:,1].reshape(-1,1), scaled_data[:,0].reshape(-1,1), test_size = 0.2, random_state = 40)
```

<font color='darkred'>Αρχικά θα εφαρμόσουμε το μοντέλο της Γραμμικής Παλινδρόμησης στα δεδομένα με χρήση της βιβλιοθήκης Scikit-Learn. Η Γραμμική Παλινδρόμηση προσαρμόζει μία ευθεία στα δεδομένα ελαχιστοποιώντας το μέσο τετραγωνικό σφάλμα  (mean squared error), το οποίο δίνεται από την σχέση:
$$ 
MSE = \frac{1}{n}\sum_{i=0}^n (y_i - \hat{y}_i )^2
$$ 
όπου $y_i$ η πραγματική τιμή και $\hat{y}_i$ η πρόβλεψη.

<font color='darkred'>Στο τέλος τυπώνονται οι συντελεστές της ευθείας γραμμικής παλινδρόμησης, η οποία είναι της μορφής $y = wx + b$ και το μέσο τετραγωνικό σφάλμα των προβλέψεων πάνω στο test set.

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression().fit(Xtrain, Ytrain)
print('Συντελεστής w =', model.coef_)
print('Συντελεστής b (σταθερό) =', model.intercept_)
predictions = model.predict(Xtest)
print('MSE στο test set = ', mean_squared_error(predictions,Ytest) )
#print('Accuracy score = ',model.score(predictions,Ytest)) #αν θέλαμε να δούμε και την ακρίβεια
```

<font color='darkred'>Παρακάτω αντί να εφαρμόσουμε το μοντέλο Γραμμικής Παλινδρόμησης έτοιμο από την βιβλιοθήκη Scikit-Learn θα υπολογίσουμε τις παραμέτρους του μοντέλου (w και b) χρησιμοποιώντας την μέθοδο Stochastic Gradient Descent (SGD). Στην αρχή εμείς ορίζουμε το ρυθμό μάθησης (learning rate) και το συνολικό αριθμό των εποχών (epochs) εκπαίδευσης. Επιπλέον, οι αρχικές τιμές των παραμέτρων w και b είναι μηδενικές.

<font color='darkred'>Στο τέλος τυπώνονται ξανά οι συντελεστές της ευθείας γραμμικής παλινδρόμησης και το μέσο τετραγωνικό σφάλμα των προβλέψεων πάνω στο test set.

```
from sklearn.utils import shuffle
learning_rate = 0.03 #εμείς το ορίζουμε
epochs = 10 #εμείς το ορίζουμε
w, b = 0.0, 0.0
mse_train, mse_test = [],[]
for epoch in range(epochs):
    X_shuffle, y_shuffle = shuffle(Xtrain, Ytrain, random_state=30) #Στοχαστική 
    for x, y in zip(X_shuffle, y_shuffle):
        d = y - (w*x + b)
        w -= learning_rate*(-2.0*x*d)
        b -= learning_rate*(-2.0*d)
    mse_train.append(mean_squared_error(w*Xtrain+b,Ytrain))
    mse_test.append(mean_squared_error(w*Xtest+b,Ytest))
 
print('Συντελεστής w =', w)
print('Συντελεστής b (σταθερό) =', b)
print('MSE στο test set = ', mean_squared_error(w*Xtest+b,Ytest))
```

<font color='darkred'>Παρατηρούμε πως με ρυθμό εκμάθησης 0.03 και αριθμό εποχών 10 ο αλγόριθμος τα πήγε πολύ καλά και έχει πολύ παρόμοια αποτελέσματα με προηγουμένως. 

<font color='darkred'>Στην συνέχεια επαναλαμβάνουμε τον κώδικα της άσκησης χρησιμοποιώντας την μέθοδο Stochastic Gradient Descent (SGD) χωρίς να πραγματοποιήσουμε κανονικοποίηση ώστε να δούμε τι θα συμβεί. 






```
df = pd.read_csv("https://raw.githubusercontent.com/nkostopoulos/StochasticsLabPublic/master/lab1/data1a.csv")
df = df[['views','likes']]
Xtrain,Xtest,Ytrain,Ytest = train_test_split(df.likes.values.reshape(-1,1), df.views.values.reshape(-1,1), test_size = 0.2, random_state = 40)
```

```
learning_rate = 0.03 #εμείς το ορίζουμε
epochs = 1000 #εμείς το ορίζουμε
w, b = 0.0, 0.0
for epoch in range(epochs):
    X_shuffle, y_shuffle = shuffle(Xtrain, Ytrain, random_state=30) #Στοχαστική 
    for x, y in zip(X_shuffle, y_shuffle):
        d = y - (w*x + b)
        w -= learning_rate*(-2.0*x*d)
        b -= learning_rate*(-2.0*d)
```

<font color='darkred'>Παρατηρούμε πως ο αλγόριθμος δεν συγκλίνει σε αυτή την περίπτωση καθώς οι τιμές των δεδομένων είναι πολύ υψηλές και κάποια στιγμή οδηγήθηκε σε απειρισμό. 

<font color='darkred'>Τέλος, θα πειραματιστούμε με τις τιμές του ρυθμού μάθησης και τον αριθμό των εποχών για να δούμε πώς αλλάζει το μέσο τετραγωνικό σφάλμα (MSE). 

Όσον αφορά τις εποχές δημιουργείται το παρακάτω διάγραμμα, το οποίο δείχνει για κάθε εποχή (έχουμε τρέξει τον αλγόριθμο για 10 εποχές) το μέσο τετραγωνικό σφάλμα στο train και στο test set. </font>


```
plt.scatter(range(1,epochs+1), mse_train, label = 'MSE train')
plt.scatter(range(1,epochs+1), mse_test, label = 'MSE test')
plt.title('Διάγραμμα MSE ανά εποχή')
plt.ylabel('MSE')
plt.xlabel('Εποχή')
plt.legend()
plt.show()
```

<font color='darkred'>Στο διάγραμμα παρατηρούμε ότι μπορούμε να επιλέξουμε έναν μικρότερο αριθμό εποχών, όπως για παράδειγμα 4, καθώς από εκεί και έπειτα το μέσο τετραγωνικό σφάλμα δεν μειώνεται ιδιαίτερα (μένει σχεδόν σταθερό). 

Τέλος, όσον αφορά το learning rate γίνονται διάφορες δοκιμές τροποποιώντας τον κώδικα ως εξής: </font>

```
mse_train, mse_test = [],[]
for leanring_rate in np.arange(0,0.2,0.01):
  epochs = 4
  w, b = 0.0, 0.0
  for epoch in range(epochs):
    X_shuffle, y_shuffle = shuffle(Xtrain, Ytrain, random_state=30) #Στοχαστική 
    for x, y in zip(X_shuffle, y_shuffle):
      d = y - (w*x + b)
      w -= learning_rate*(-2.0*x*d)
      b -= learning_rate*(-2.0*d)
  mse_train.append(mean_squared_error(w*Xtrain+b,Ytrain))
  mse_test.append(mean_squared_error(w*Xtest+b,Ytest))

print('Συντελεστής w =', w)
print('Συντελεστής b (σταθερό) =', b)
print('MSE στο test set = ', mean_squared_error(w*Xtest+b,Ytest))
```

<font color='darkred'>Δημιουργούμε το διάγραμμα όπου στον x άξονα έχει το learning rate και στον y το MSE.

```
plt.scatter(np.arange(0,0.2,0.01), mse_train, label = 'MSE train')
plt.scatter(np.arange(0,0.2,0.01), mse_test, label = 'MSE test')
plt.title('Διάγραμμα MSE ανά learning rate')
plt.ylabel('MSE')
plt.xlabel('Learrning rate')
plt.legend()
plt.show()
```

<font color='darkred'>Παρατηρούμε πως με οποιοδήποτε learning rate δοκιμάσαμε (στο διάστημα [0,0.2] με βήμα 0.01) το μέσο τετραγωνικό σφάλμα βγαίνει σχεδόν ίδιο και ο αλγόριθμος συγκλίνει γρήγορα.

<font color='darkred'>Στην συνέχεια υπολογίζουμε τις παραμέτρους του μοντέλου (w και b) χρησιμοποιώντας την μέθοδο Stochastic Gradient Descent (SGD)και τυπώνουμε ξανά τους συντελεστές της ευθείας γραμμικής παλινδρόμησης και το μέσο τετραγωνικό σφάλμα των προβλέψεων πάνω στο test set. Παρατηρούμε πως όπως και στο προηγούμενο παράδειγμα ο αλγόριθμος τα πηγαίνει πολύ καλά με ρυθμό εκμάθησης 0.03 και αριθμό εποχών 10, έχοντας παρόμοια αποτελέσματα. 


```
learning_rate = 0.03 #εμείς το ορίζουμε
epochs = 10 #εμείς το ορίζουμε
w, b = 0.0, 0.0
mse_train, mse_test = [],[]
for epoch in range(epochs):
    X_shuffle, y_shuffle = shuffle(Xtrain, Ytrain, random_state=30) #Στοχαστική 
    for x, y in zip(X_shuffle, y_shuffle):
        d = y - (w*x + b)
        w -= learning_rate*(-2.0*x*d)
        b -= learning_rate*(-2.0*d)
    mse_train.append(mean_squared_error(w*Xtrain+b,Ytrain))
    mse_test.append(mean_squared_error(w*Xtest+b,Ytest))

print('Συντελεστής w =', w)
print('Συντελεστής b (σταθερό) =', b)
print('MSE στο test set = ', mean_squared_error(w*Xtest+b,Ytest))
```

```
df2 = pd.read_csv("https://raw.githubusercontent.com/nkostopoulos/StochasticsLabPublic/master/lab1/data1b.csv")
Xtrain,Xtest,Ytrain,Ytest = train_test_split(df2.x.values.reshape(-1,1), df2.y.values.reshape(-1,1), test_size = 0.2, random_state = 40)
```

Τέλος, όσον αφορά το learning rate γίνονται διάφορες δοκιμές τροποποιώντας τον κώδικα ως εξής: </font>


```
mse_train, mse_test = [],[]
for leanring_rate in np.arange(0,0.2,0.01):
  epochs = 2
  w, b = 0.0, 0.0
  for epoch in range(epochs):
    X_shuffle, y_shuffle = shuffle(Xtrain, Ytrain, random_state=30) #Στοχαστική 
    for x, y in zip(X_shuffle, y_shuffle):
      d = y - (w*x + b)
      w -= learning_rate*(-2.0*x*d)
      b -= learning_rate*(-2.0*d)
  mse_train.append(mean_squared_error(w*Xtrain+b,Ytrain))
  mse_test.append(mean_squared_error(w*Xtest+b,Ytest))

print('Συντελεστής w =', w)
print('Συντελεστής b (σταθερό) =', b)
print('MSE στο test set = ', mean_squared_error(w*Xtest+b,Ytest))
```
