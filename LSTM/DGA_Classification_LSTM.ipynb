Χρησιμοποιείται ένας μεγάλος αριθμός από τέτοια ονόματα, η πλειοψηφία των οποίων δεν έχουν κάποια αντιστοίχιση σε διεύθυνση IP και στοχεύουν στην απόκρυψη του Command & Control Server, επειδή οι αμυνόμενοι καλούνται να ελέγξουν κάθε ένα από τα ονόματα που παρατηρούν στο δίκτυό τους, σπαταλώντας χρόνο και πόρους. Επιπρόσθετα, η διεύθυνση IP του Command & Control Server αλλάζει πολύ συχνά (πολλές φορές σε μία μέρα), ώστε να αποφεύγεται ο εντοπισμός του ακόμα και όταν εντοπίζονται τα ονόματα DGA που οδήγησαν σε αυτόν.

```python
import numpy
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.layers import Embedding
from keras.layers import LSTM
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt
import random
import itertools
import seaborn as sns

def add_word_distinct_chars(string, distinct_chars):
    for char in string:
        distinct_chars.add(char)
    return distinct_chars

def find_max_len(string, max_len):
    string_length = len(string)
    if string_length > max_len:
        max_len = string_length
    return max_len

def load_data(filename):
    dataset = []
    distinct_chars = set()
    max_len = 0
    currentIndex = 0
    diverse_labels = dict()
    with open(filename, "r") as fdr:
        for line in fdr:
            line = line.strip()
            general, label, name = line.split(",")
            name = name.split(".")[0]
            if label not in diverse_labels.keys():
                diverse_labels[label] = currentIndex
                currentIndex += 1
            distinct_chars = add_word_distinct_chars(name, distinct_chars)
            max_len = find_max_len(name, max_len)
            temp_list = []
            temp_list.append(name)
            temp_list.append(label)
            dataset.append(temp_list)
    random.shuffle(dataset)
    return dataset, distinct_chars, max_len, diverse_labels

def assign_index(chars):
    features = {}
    for index, char in enumerate(chars):
        features[char] = index
    return features

def convert_dataset_and_tokenize(dataset, features, max_len):
    for item_no, example in enumerate(dataset):
        name = example[0]
        label = example[1]
        tokenized = []
        padding_needed = max_len - len(name)
        for index in range(padding_needed):
            tokenized.append(0)
        for char in name:
            token = features[char]
            tokenized.append(token)
        example[0] = tokenized
        dataset[item_no] = example
    return dataset

def split_examples_labels(dataset):
    examples = [entry[0] for entry in dataset]
    labels = [entry[1] for entry in dataset]
    return examples, labels

def convert_labels_to_int(labels, diverse_labels):
    for index, label in enumerate(labels):
        labels[index] = diverse_labels[label]
    return labels

def build_model(max_features, max_len):
    model = Sequential()
    model.add(Embedding(max_features, 128, input_length = max_len))
    model.add(LSTM(128))
    model.add(Dropout(0.5))
    model.add(Dense(26, activation = 'softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    return model

def return_fold(examples, labels_int, num_folds, current_fold):
    interval = len(examples) // num_folds
    current_start = (current_fold - 1) * interval
    current_end = current_fold * interval
    X_test = examples[current_start:current_end]
    y_test = labels_int[current_start:current_end]
    X_train = examples[:current_start]
    y_train = labels_int[:current_start]
    X_train.extend(examples[current_end:])
    y_train.extend(labels_int[current_end:])
    return X_train, y_train, X_test, y_test

def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    y_pred = numpy.argmax(y_pred, axis = 1)
    return y_pred

def return_classification_report(y_test, y_pred, diverse_names):
    target_names = []
    for item in diverse_names.keys():
        target_names.append(item)
    print(classification_report(y_test, y_pred, target_names = target_names))
    return None
    
if __name__ == "__main__":
    dataset, distinct_chars, max_len, diverse_labels = load_data("dga_domains_full.csv")
    dataset = dataset[0:100000]
    max_features = len(distinct_chars) + 1
    features = assign_index(distinct_chars)
    dataset = convert_dataset_and_tokenize(dataset, features, max_len)
    examples, labels = split_examples_labels(dataset)
    labels_int = convert_labels_to_int(labels, diverse_labels)

    acc_per_fold = []
    loss_per_fold = []
    number_of_folds = 5
    for fold in range(1, number_of_folds + 1):
        X_train, y_train, X_test, y_test = return_fold(examples, labels_int, number_of_folds, fold)
        model = build_model(max_features, max_len)
        print("Training for fold: ", fold)
        history = model.fit(X_train, y_train, batch_size = 128, epochs = 5, verbose = 1)
        scores = model.evaluate(X_test, y_test, verbose = 1)
        print(f'Score for fold {fold}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
        acc_per_fold.append(scores[1] * 100)
        loss_per_fold.append(scores[0])
        y_pred = make_predictions(model, X_test)

        #better visualization of confusion matrix
        cnf = confusion_matrix(y_test, y_pred)
        fig, ax = plt.subplots(figsize=(16, 16))
        im = ax.imshow(cnf, interpolation='nearest', cmap=plt.cm.Blues)
        ax.figure.colorbar(im, ax=ax)
        ax.set(xticks=numpy.arange(cnf.shape[1]),
              yticks=numpy.arange(cnf.shape[0]),
              xticklabels=diverse_labels, yticklabels=diverse_labels,
              xlabel='Predicted',
              ylabel='Actual',
              title='Confusion Matrix')
        fmt = 'd'
        thresh = cnf.max() / 2.
        for i, j in itertools.product(range(cnf.shape[0]), range(cnf.shape[1])):
            ax.text(j, i, format(cnf[i, j], fmt), horizontalalignment="center", color="white" if cnf[i, j] > thresh else "black")

        plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
        plt.show()
        return_classification_report(y_test, y_pred, diverse_labels)

    # == Provide average scores ==
    print('------------------------------------------------------------------------')
    print('Score per fold')
    for i in range(0, len(acc_per_fold)):
        print('------------------------------------------------------------------------')
        print(f'> Fold {i+1} - Loss: {loss_per_fold[i]} - Accuracy: {acc_per_fold[i]}%')
        print('------------------------------------------------------------------------')
        print('Average scores for all folds:')
        print(f'> Accuracy: {numpy.mean(acc_per_fold)} (+- {numpy.std(acc_per_fold)})')
        print(f'> Loss: {numpy.mean(loss_per_fold)}')
        print('------------------------------------------------------------------------')
```

<font color='darkred'>Το embedding layer είναι ένα από τα διαθέσιμα layers στο Keras, το οποίο χρησιμοποιείται κυρίως σε εφαρμογές επεξεργασίας φυσικής γλώσσας (NLP), όπως η μοντελοποίηση γλώσσας. Το embedding layer μας δίνει την δυνατότητα να μετατρέψουμε κάθε λέξη σε ένα σταθερού μήκους διάνυσμα ώστε στην συνέχεια τα διανύσματα να εισαχθούν στο δίκτυο LSTM. Το διανύσματα που προκύπτουν είναι πυκνά (dense) με πραγματικές τιμές. Το σταθερό μήκος των διανυσμάτων μας βοηθά να αναπαραστήσουμε τις λέξεις με καλύτερο τρόπο με μειωμένες διαστάσεις. Με αυτόν τον τρόπο το embedding layer λειτουργεί σαν πίνακας αναζήτησης, όπου οι λέξεις είναι τα κλειδιά σε αυτόν τον πίνακα, ενώ τα πυκνά διανύσματα είναι οι τιμές. Στην συγκεκριμένη υλοποίηση το embedding layer δέχεται τρεις παραμέτρους, οι οποίες είναι οι εξής:
* Input dim = Vocabulary size (max features), δηλαδή το πλήθος των διαφορετικών χαρακτήρων που εμφανίζονται στα δεδομένα συν 1
*  Output dim = 128, το οποίο είναι το μήκος των πυκνών διανυσμάτων
*  Input length = maximum length, το οποίο είναι το μήκος των ακολουθιών. Σε αυτήν την υλοποίηση βάζουμε σαν τιμή το μέγεθος της μεγαλύτερης συμβολοσειράς που παρατηρείται στα δεδομένα μας </font>

<li>Να αναφέρετε λόγους για τους οποίους χρησιμοποιείται η μέθοδος K-fold cross validation.</li>

<font color='darkred'>Η μέθοδος k-fold cross validation χρησιμοποιείται για την αξιολόγηση μοντέλων μηχανικής μάθησης. Το κύριο πλεονέκτημα της μεθόδου είναι ότι αποφεύγεται το overfitting στα δεδομένα εκπαίδευσης καθώς κάθε φορά το σύνολο εκπαίδευσης διαφέρει. Με αυτό τον τρόπο επιτυγχάνεται και καλύτερη προσέγγιση στην πραγματική απόδοση του μοντέλου, όσον αφορά τις μετρικές που χρησιμοποιούνται(για παράδειγμα accuracy). Αν χρησιμοποιούταν ένα απλό train-test split τότε η απόδοση του μοντέλου μπορεί να τύχαινε να είναι πολύ υψηλή λόγω της συγκεκριμένης κατανομής των δεδομένων, ενώ με την χρήση k-fold CV αυτό αποτρέπεται. Τέλος, είναι σημαντικό να αναφερθεί πως η μέθοδος αυτή εμφανίζει καλά αποτελέσματα και σε μικρά datasets, καθώς επιτρέπει όλα τα δεδομένα να χρησιμοποιηθούν εν γένει και στην εκπαίδευση και στην αξιολόγηση του μοντέλου. 

<li>Κατά τη χρήση της μεθόδου K-fold cross validation, ανά fold, να αναφέρετε πόσα examples χρησιμοποιούνται για την εκπαίδευση του μοντέλου και πόσα για το validation/testing.

<font color='darkred'>Στο Cross Validation αρχικά χωρίζουμε το training set σε έναν αριθμό "πτυχών" (folds). Συνηθισμένες τιμές είναι το 5 και το 10 (5-fold και 10-fold CV). Στη συνέχεια, για κάθε k-fold θεωρούμε ότι τα k μείον 1 folds είναι training set και ότι το fold που αφήσαμε έξω είναι το test set (Προφανώς κάθε fold έχει ίσο αριθμό δειγματικών στοιχείων). Στην συγκεκριμένη άσκηση έχουμε 5-fold CV (k=5) και το dataset περιέχει N = 100.000 δείγματα (ονόματα DNS). Συνεπώς κάθε φορά για εκπαίδευση χρησιμοποιούνται 80.000 δείγματα και για αξιολόγηση 20.000 δείγματα.

<font color='darkred'> Πέρα από K-fold cross validation μπορούν να χρησιμοποιηθούν διάφορες μέθοδοι. Παρακάτω αναλύω μερικές από αυτές:
*   **Holdout cross validation (train-test split)**: Τα δεδομένα χωρίζονται με τυχαίο τρόπο σε δεδομένα training και δεδομένα testing σε κάποια αναλογία, η οποία είναι συνήθως 80:20 ή 70:30. Η μέθοδος αυτή αποτελεί την συνηθισμένη πρακτική όταν εκπαιδεύουμε ένα μοντέλο. Ωστόσο, η συγκεκριμένη μέθοδος έχει ορισμένα μειονεκτήματα, όπως το ότι μπορεί τα δεδομένα training να μην είναι αντιπροσωπευτικά για όλο το δείγμα και ότι τα δεδομένα testing να έχουν χρήσιμη πληροφορία, η οποία θα χαθεί. Όπως ανέφερα προηγουμένως το κύριο πλεονέκτημα του k-fold cross validation είναι η αποφυγή του overfitting, η οποία μπορεί να πραγματοποιηθεί σε ένα απλό train-test split.
*   **Stratified k-fold cross validation**: Παρόμοια με το k-fold cross validation με την διαφορά ότι κάθε fold αντιπροσωπεύει το σύνολο των δεδομένων. Η μέθοδος αυτή είναι ιδιαίτερα σημαντική σε προβλήματα που αφορούν σε μεγάλο βαθμό imbalanced κλάσεις.
*  **Leave-p-out cross validation**: Από τα συνολικά διαθέσιμα δεδομένα, έστω N, «αφήνουμε έξω» p στοιχεία, τα οποία χρησιμοποιούνται για αξιολόγηση του μοντέλου και τα υπόλοιπα N-p χρησιμοποιούνται για την εκπαίδευση του. Αυτό επαναλαμβάνεται για όλους του δυνατούς συνδυασμούς. Πλεονέκτημα της μεθόδου είναι ότι προσεγγίζει καλύτερα την απόδοση του μοντέλου, ειδικότερα όταν πρόκειται για μικρά και πολύπλοκα datasets, ενώ αποφεύγεται το overfitting. Προφανώς το κύριο μειονέκτημα είναι οτι είναι πολύ ακριβή μέθοδος υπολογιστικά.
* **Leave-1-out cross validation**: Ειδική περίπτωση της Leave-p-out cross validation με p = 1, το οποίο σημαίνει πως κάθε φορά το μοντέλο εκπαιδεύεται σε όλα τα δεδομένα εκτός από μόνο ένα στοιχείο, το οποίο χρησιμοποιείται για αξιολόγηση. Μοιράζεται τα ίδια πλεονεκτήματα με την Leave-p-out cross validation. Επιπλέον, αξίζει να αναφερθεί πως σε μη ισορροπημένα datasets, όπου μία κλάση μπορεί να επικρατεί σε μεγάλο βαθμό έναντι των υπολοίπων, δεν είναι κατάλληλη μέθοδος. </font> 

<li>Να αναλύσετε τα κριτήρια precision, recall και F1-score που εμφανίζονται στο classification report.</li>

<font color='darkred'>Τα κριτήρια αυτά έχουν αναλυθεί με λεπτομέρεια στο lab 10 (Decision Trees and Random Forests), καθώς και για ποιους λόγους είναι απαραίτητα σε πλήθος προβλημάτων πέραν της ακρίβειας (accuracy). Μία συνοπτική παρουσίαση των κριτηρίων φαίνεται και παρακάτω: 
$Precision (P)$ είναι ο λόγος των true positives ($T_p$) ως προς τον αριθμό των true positives συν τον αριθμό των false positives ($F_p$):
$$Precision = \frac{T_p}{T_p+F_p}$$
$Recall(R)$ είναι ο λόγος των true positives ($T_p$) ως προς τον αριθμό των true positives συν τον αριθμό των false negatives ($F_n$):
$$Recall = \frac{T_p}{T_p + F_n}$$
$F_1-score$ είναι ο αρμονικός μέσος της ακρίβειας και της ανάκλησης. Συγκεκριμένα:
$$F1-score = 2\frac{P \times R}{P+R}$$
